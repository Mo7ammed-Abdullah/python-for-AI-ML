```python
# convert list,tuple to dataframe

import pandas as pd
my_list = [["Alice", 25], ["Bob", 30], ["Charlie", 35 ]]
df = pd.DataFrame(my_list)

print(type(df))
print(df)  # print the dataframe

# change column name
df1 = pd.DataFrame(my_list, columns=["Name", "Age"])
print(df1)

# provide index
df2 = pd.DataFrame(my_list, columns=["Name", "Age"], index=[1,2,3])
print(df2)


# convert dictionary to dataframe
my_dict = {
    "Name": ["Alice", "Bob", "Charlie"], 
    "Age": [25, 30, 35]
    }
df3 = pd.DataFrame(my_dict)
print(df3)


# convert list of dictionary to dataframe
my_list = [
    {"Name": "Alice", "Age": 25 , "City": "New York"}, 
    {"Name": "Bob", "Age": 30 , "City": "London"}, 
    {"Name": "Charlie", "Age": 35}
    ]
df4 = pd.DataFrame(my_list)
print(df4)
```

```python
# access data at a specific location in the DataFrame , at specific col and specific row
# df.loc[index_start:index_end, col_start:col_end]
import pandas as pd

# display all values at 0 number index
df = pd.read_csv('student_data.csv')
print(df.loc[0])

# display multiple rows
print([df.loc[[2,3,19]]])   #list of index of all the rows that i wanna see


# display multiple rows in a range
print(df.loc[3:10])   #index from 3 to 10


# display all rows of a specific column
print(df.loc[:, "Python Marks"])


# display all rows of multiple columns
print(df.loc[:, ["Python Marks", "Algorithm Marks"]])


# display row and clomns with in a range
print(df.loc[3:10, "CompletionStatus":"Instructor"])
```

```python
# iterate through rows , using the below loop is slow
import pandas as pd
df = pd.read_csv('student_data.csv')

for index, row in df.iterrows():
    print(f"{index}:{row}")
```

```python
# iterate through rows , using tuple which is a bit faster
import pandas as pd
df = pd.read_csv('student_data.csv')

for i in df.itertuples():
    print(i)
```

```python
# sorting based on data of a column

import pandas as pd
df = pd.read_csv('student_data.csv')

# sort in ascending order
copy = df.sort_values("Data Structure Marks")
print(copy)

# sort in descending order
copy = df.sort_values("Data Structure Marks", ascending=False)
print(copy)

# sort multiple columns , first sort based on Data Structure Marks and if any student has same marks in Data Structure Marks then sort based on Python Marks
copy = df.sort_values(["Data Structure Marks", "Python Marks"])
print(copy)

# sorting first column on descending and second column on ascending if any student has same marks in Data Structure Marks
copy = df.sort_values(["Data Structure Marks", "Python Marks"], ascending=[False, True])
print(copy)
```

```python
# filtering data based on conditions
import pandas as pd
df = pd.read_csv('student_data.csv')

# filter out data of those students that did not start the course in completion status column
copy = df.loc[df["CompletionStatus"] == "Not Started"]
print(copy)


# filter out students that completed course and have data structure marks greater than 90
copy = df.loc[(df["CompletionStatus"] == "Completed") & (df["Data Structure Marks"] >= 90)]
print(copy)
```

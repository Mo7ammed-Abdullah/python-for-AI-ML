# Seaborn and plotly

seaborn: A data visualization library built on top of Matplotlib that provides a high-level interface for creating attractive and informative statistical graphics

plotly: Plotly's primary strength lies in creating interactive plots that can be embedded in web applications, dashboards, or shared as standalone HTML files.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
student = pd.read_csv('sns_data.csv') 
student.head()
```

# 1. line plot using seaborn


```python
#axes level (a single subplot)
sns.lineplot(data= student , x = 'week' , y = 'attendance_rate',errorbar=None)
```

```python
# axes level
# line plot based on gender
sns.lineplot(data= student , x = 'week' , y = 'attendance_rate',errorbar=None,hue='gender')
```

```python
# figure level (the window containing plots)
sns.relplot(kind ='line', data= student , x = 'week' , y = 'attendance_rate',errorbar=None,hue='gender')

# large figure
```

```python
sns.relplot(kind ='line', data= student , x = 'week' , y = 'test_score',errorbar=None,hue='class_level')
```

# 5. KDE is seaborn

Kernel Density Estimate, which is a function used to visualize the probability density of a continuous variable. 

It creates a smooth, continuous curve that represents the distribution of data

```python
# axes level
# density of attendance rate
sns.kdeplot(data = student , x ='attendance_rate') 
```

```python
# lets create a Kde plot for this tip
sns.displot(kind='hist' , data = tips_data, x ='tip', bins=20) 
```

```python
# the kde plot for above histogram
sns.kdeplot(data = tips_data, x ='tip') 
```

```python
# based on gender
sns.kdeplot(data = tips_data, x ='tip',hue='sex',fill=True) 
```

```python

#figure level

sns.displot(kind='kde' , data = student , x ='attendance_rate',col='gender')
```

# 7. Bar plot
barplot() is used to visualize a numerical variable's aggregate or statistical estimate (e.g., mean, median, sum) across different categories of a categorical variable.

```python
# will show the mean of marks based on gender by default
sns.barplot(data = student , x = 'gender',y = 'Marks',errorbar=None) 
```

# 8. Regplot
regplot is a function in the Python library Seaborn that plots a scatter plot of data and fits a linear regression model to it. 

It is used to visualize the linear relationship between two continuous variables

```python
# axes level
sns.regplot(data = student , x ='study_hours' , y = 'test_score') 
```

```python
#figure level 

sns.lmplot(data = student , x ='study_hours' , y = 'test_score',hue='gender') 
```

# 9. pairplot
A pair plot, also known as a scatterplot matrix, is a data visualization that shows pairwise relationships between multiple variables in a dataset.

It creates a grid of plots where each variable is compared against every other variable using scatter plots, while the diagonal shows the distribution of each individual variable, typically with histograms or kernel density plots. 

Pair plots are useful for quickly identifying patterns, correlations, and trends across your data. 

```python
student_marks = student[['Marks','study_hours','attendance_rate','gender']]
student_marks
```

```python
sns.pairplot(student_marks)
```

```python
sns.pairplot(data=student_marks,hue='gender')
```

```python
sns.pairplot(data=student_marks, kind='hist')
```

# 10. joint plot

A joint plot is a visualization that combines a scatter plot of two variables with their individual distributions, which are typically displayed as histograms or kernel density estimates (KDEs) along the top and right margins. 

This allows you to see both the relationship between two variables and their individual distributions on a single figure. 

```python
sns.jointplot(data=student_marks , x ='study_hours' , y='Marks') 
```

```python
sns.jointplot(data=student_marks , x ='study_hours' , y='Marks',hue='gender') 
```

```python
sns.jointplot(data=student_marks , x ='study_hours' , y='Marks',kind='kde') 
```

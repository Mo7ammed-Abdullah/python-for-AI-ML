# Matplotlib is used for data visualization

```python
import numpy as pd
import pandas as pd
import matplotlib.pyplot as plt
```

# 1. 2D line plot
A graphical representation of data where data points are connected by lines
to show how a value changes over time

Can use plt.show() to display the plot in every platform

```python
# 2D line plotting - plot(x , y)

hours = [2 , 3 , 4 , 1 , 5 , 7 , 3]
days = [1 , 2 , 3 , 4 , 5 , 6 , 7]
plt.plot(days , hours)
```

```python
# labeling x and y axis
plt.xlabel("Days")
plt.ylabel("Hours Studied")
plt.plot(days , hours)
```

```python
# title to the graph
plt.title("Trend of student study time over a week")
plt.xlabel("Days")
plt.ylabel("Hours Studied")
plt.plot(days , hours)
```

```python
# multiple plots in a graph
hours_1 = [2 , 3 , 4, 1 , 5 , 7 , 3] 
hours_2 = [ 1 , 4 , 3 , 5 ,7 ,2 ,1]
days = [1 ,2 , 3 ,4 ,5 ,6 ,7] 

plt.xlabel('Days')
plt.ylabel('Study Hours') 
plt.title('Trend of a student study time over a week')

plt.plot(days,hours_1)
plt.plot(days,hours_2) 
```

# 2. 2d line plot from pandas dataset

```python
df = pd.read_csv('enrollment_data.csv')
df
```

```python
# plot with the enrollment data , x axis have year and y axis have programming and digital marketing
plt.xlabel('Year')
plt.ylabel('Count of enrollment')
plt.title('Student enrollment over 8 years')
plt.plot(df['Year'] , df['Programming'])
plt.plot(df['Year'] , df['Digital Marketing'])
```

```python
# adding labels to plots using plt.legend()
plt.xlabel('Years') 
plt.ylabel('Count of Enrollment')
plt.title('Student Enrollment over 8 years')

plt.plot(df['Year'] ,df['Programming'],label = 'Programming')
plt.plot(df['Year'] ,df['Digital Marketing'], label= 'Digital Marketing')

plt.legend()   #  to change location plt.legend(loc = 'upper left')
```

# 3. Customizing Line plot ( color,linestyle,marker and Grid )

```python
# change color of plot
plt.xlabel('Years') 
plt.ylabel('Count of Enrollment')
plt.title('Student Enrollment over 8 years')

plt.plot(df['Year'] ,df['Programming'],label = 'Programming' , color = 'red')
plt.plot(df['Year'] ,df['Digital Marketing'], label= 'Digital Marketing' , color = 'black')

plt.legend() 
```

```python
# increase line width
plt.xlabel('Years') 
plt.ylabel('Count of Enrollment')
plt.title('Student Enrollment over 8 years')

plt.plot(df['Year'] ,df['Programming'],label = 'Programming' , color = 'red' , linewidth = 5)
plt.plot(df['Year'] ,df['Digital Marketing'], label= 'Digital Marketing' , color = 'black' , linewidth = 5)

plt.legend()
```

```python
# line style
plt.xlabel('Years') 
plt.ylabel('Count of Enrollment')
plt.title('Student Enrollment over 8 years')

plt.plot(df['Year'] ,df['Programming'],label = 'Programming' , color = 'red' , linewidth = 3 , linestyle = 'dashed')
plt.plot(df['Year'] ,df['Digital Marketing'], label= 'Digital Marketing' , color = 'black' , linewidth = 3 , linestyle = 'dotted')

plt.legend()
```

```python
# add points to plot
plt.xlabel('Years') 
plt.ylabel('Count of Enrollment')
plt.title('Student Enrollment over 8 years')

plt.plot(df['Year'] ,df['Programming'],label = 'Programming' , color = 'red' , linewidth = 2 , linestyle = 'dashed' , marker = 'o')
plt.plot(df['Year'] ,df['Digital Marketing'], label= 'Digital Marketing' , color = 'black' , linewidth = 2 , linestyle = 'dotted' , marker = 'x')

plt.legend()
```

```python
# add grid to the graph
plt.xlabel('Years') 
plt.ylabel('Count of Enrollment')
plt.title('Student Enrollment over 8 years')

plt.plot(df['Year'] ,df['Programming'],label = 'Programming' , color = 'red' , linewidth = 2 , linestyle = 'dashed' , marker = 'o')
plt.plot(df['Year'] ,df['Digital Marketing'], label= 'Digital Marketing' , color = 'black' , linewidth = 2 , linestyle = 'dotted' , marker = 'x')

plt.legend()
plt.grid()
```

# 4. Scatter plots
to find correlation between datas to train a model

```python
students = pd.read_csv('student_IQdata.csv')
students
```

```python
#Relation between Study Hours and IQ Score - postive correlation

plt.xlabel('Study Hours') 
plt.ylabel('IQ Score') 
plt.title('Relation between Study Hours and IQ Score') 

plt.scatter(students['Study_Hour'] , students['IQ_Score']) 

# the farest points are the outliers which may make the model biased to the outliers
```

```python
#Relation between Shoe size and IQ Score - > no correlation

plt.xlabel('Shoe Size') 
plt.ylabel('IQ Score') 
plt.title('Relation between Shoe Size and IQ Score') 

plt.scatter(students['Shoe_Size'] , students['IQ_Score']) 
```

```python
#Relation between Chilling hour and IQ Score - > negative correlation
plt.xlabel('Chilling Time') 
plt.ylabel('IQ Score') 
plt.title('Relation between Chilling hours and IQ Score') 

plt.scatter(students['Chilling_Hours'] , students['IQ_Score'],marker='o',color='red') 
```

# 5. Histogram


```python
# based on iq , here y is not given so it will show frequency of students (how many students have this iq score)

plt.hist(students['IQ_Score'],color='Red',edgecolor='black')
plt.xlabel('IQ Score') 
plt.ylabel('Frequency of Students') 

# 140 IQ score is an outlier
```

```python
# based on study hours and custom bins

plt.hist(students['Study_Hour'],bins=15,color='Red',edgecolor='black') # bins are number of bars
plt.xlabel('Study Hours') 
plt.ylabel('Frequency of Students') 
```

```python
# bins of 1 hour

plt.hist(students['Study_Hour'],bins=[1,2,3,4,5,6,7,8],color='Red',edgecolor='black')
plt.xlabel('Study Hours') 
plt.ylabel('Frequency of Students') 
```

# 6. Bar chart

```python
data = {
    'Student': [ 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Gender': [ 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'StudyHours': [ 2, 5, 3, 7, 4, 8, 1],
    'Attendance': [ 60, 88, 70, 95, 80, 98, 55],
    'Grade': [ 'Fail', 'Pass', 'Fail', 'Pass', 'Pass', 'Pass', 'Fail']
}

df = pd.DataFrame(data)
df
```

```python
# group by gender and count male and female

gender_group = df.groupby('Gender').size() 

print(gender_group.index)
gender_group.values
```

```python
# create bar chart of index and values of gender group

plt.bar(gender_group.index,gender_group.values)
```

```python
# customize

plt.bar(gender_group.index,gender_group.values,color=['orange','Red'],edgecolor='black')

plt.xlabel('Gender') 
plt.ylabel('Count') 
plt.title('Bar Chart of Gender among students')
```

```python
#  Bar chart on real data 

df = pd.read_csv('student_data.csv') 

df
```

```python
status = df.groupby('CompletionStatus').size() 

plt.bar(status.index,status.values,color=['green','orange','red'],edgecolor='black')

plt.xlabel('Completion Status') 
plt.ylabel('Count') 
plt.title('Bar Chart of Course completion')
plt.show() 
```

```python
# location 

status = df.groupby('Location').size() 

plt.bar(status.index,status.values,edgecolor='black')

plt.xlabel('Location') 
plt.ylabel('Count') 
plt.title('Bar Chart of Location of Students')
plt.show() 
```

# 7. Pie chart
Shows on 100% percentage

```python
# completion status

status = df.groupby('CompletionStatus').size() 

status
```

```python
# pie chart on completion status

plt.pie(status)
plt.show()
```

```python
# labeling

plt.pie(status,labels=status.index)
plt.show()
```

```python
# show percentage

plt.pie(status,labels=status.index,autopct='%1.1f%%') # copy the autopct from matplotlib documentation
plt.show()
```

```python
# other designings available in matplotlib documentation
plt.pie(status,labels=status.index,autopct='%1.1f%%',explode = (0.1,0,0),shadow=True) 
plt.show() 
```
